```python
import findspark
findspark.init('/home/ubuntu/spark-3.2.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('infosys722_step2').getOrCreate()
```

```python
dataset = spark.read.load("./Waste Water Treatment - Reformatted.csv", format="csv", header="true")
```

```python
dataset = dataset.select('Variable', 'Country', 'Year', 'PercentageValue', 'Water_quality_T', 'Life_satisfaction_T')
dataset.show()
```

```python
print("4.2 Data Projection")
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

variableArr = np.array(dataset.select('Variable').collect())
fig, ax = plt.subplots()
fig.autofmt_xdate()

plt.hist(variableArr)
plt.show()
```

```python
yearArr = np.array(dataset.select('Year').collect())
fig, ax = plt.subplots()
fig.autofmt_xdate()

plt.hist(yearArr)
plt.show()
```

```python
coutryArr = np.array(dataset.select('Country').collect())
fig, ax = plt.subplots()
fig.autofmt_xdate(rotation=90)

plt.hist(coutryArr)
plt.show()
```

```python
dataset.write.option("header",True).csv('Waste Water Treatment - Transformated.csv')
```
